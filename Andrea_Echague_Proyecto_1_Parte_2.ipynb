```python
# 1. Preparación del entorno
# Importar librería
import pandas as pd

# 2. Cargar los datos
# cargar el archivo CSV en un DataFrame
df = pd.read_csv(r"C:\Users\Andy\Documents\project1-retail-analytics\retail_sales_dataset.csv")

# 3. Exploración inicial de los datos
# Mostrar las primeras 10 filas
print("=== Primeras 10 filas del DataFrame ===")
print(df.head(10), "\n")

# Mostrar las últimas 5 filas
print("=== Últimas 5 filas del DataFrame ===")
print(df.tail(5), "\n")

# Obtener información general sobre el DataFrame
print("=== Información general del DataFrame ===")
df.info()
print("\n")

# Generar estadísticas descriptivas
print("=== Estadísticas descriptivas del DataFrame ===")
print(df.describe(), "\n")

# 4. Inspección de los datos
# Inspeccionar tipos de datos
print("=== Tipos de datos por columna ===")
print(df.dtypes, "\n")

# Contar valores únicos en la columna Product Category
print("=== Conteo de valores únicos en 'Product Category' ===")
print(df['Product Category'].value_counts(), "\n")

# Mostrar valores únicos en la columna Customer ID
print("=== Valores únicos en 'Customer ID' ===")
print(df['Customer ID'].unique()[:10], "\n")  # Mostrando solo los primeros 10 para brevedad

# 5. Filtrado de Datos
# Filtrar donde Total Amount > 50
df_total_amount = df[df['Total Amount'] > 50]
print("=== Filtrado: 'Total Amount' > 50 ===")
print(df_total_amount.head(), "\n")

# Filtrar donde Price per Unit < 0.5
df_price = df[df['Price per Unit'] < 0.5]
print("=== Filtrado: 'Price per Unit' < 0.5 ===")
print(df_price.head(), "\n")

# Filtrar con query para categoría Manzanas y Total Amount > 30
df_beauty_sales = df.query("`Product Category` == 'Manzanas' and `Total Amount` > 30")
print("=== Filtrado con query: 'Manzanas' y 'Total Amount' > 30 ===")
print(df_beauty_sales.head(), "\n")

# Filtrar con query para categoría Beauty (ya que no hay 'Manzanas') y Total Amount > 30
df_beauty_sales = df.query("`Product Category` == 'Beauty' and `Total Amount` > 30")
print("=== Filtrado con query: 'Beauty' y 'Total Amount' > 30 ===")
print(df_beauty_sales.head(), "\n")

# 6. Slicing de datos
# Seleccionar columnas Product Category y Total Amount
df_product_sales = df[['Product Category', 'Total Amount']]
print("=== Columnas 'Product Category' y 'Total Amount' ===")
print(df_product_sales.head(), "\n")

# Usar loc[] para filas 5 a 10 y columnas Product Category y Customer ID
df_loc_selection = df.loc[5:10, ['Product Category', 'Customer ID']]
print("=== Filas 5 a 10 para las columnas 'Product Category' y 'Customer ID' ===")
print(df_loc_selection, "\n")

# Usar iloc[] para primeras 5 filas y primeras 3 columnas
df_iloc_selection = df.iloc[:5, :3]
print("=== Primeras 5 filas y primeras 3 columnas ===")
print(df_iloc_selection, "\n")
```
